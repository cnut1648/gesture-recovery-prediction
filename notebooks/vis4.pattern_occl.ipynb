```python
with open("/home/jiashu/seq/avg_Lpattern[k=1].pkl", "rb") as f:
    f = pickle.load(f)

with open("/home/jiashu/seq/avg_Rpattern[k=1].pkl", "rb") as g:
    g = pickle.load(g)

jsond = defaultdict(lambda : defaultdict(lambda: defaultdict(list)))
assert set(f.keys()) == set(g.keys())
for patt in f:
    pattern = list(patt)
    for label in [0, 1]:
        for LR, fg in zip(["L", "R"], [f, g]):

            for case, delta in fg[patt][f"label {label}"]:
                if delta > 0:
                    delta = 1
                elif delta < 0:
                    delta = -1
                if delta == 0: continue
                jsond[str(pattern)][f"label {label}"][LR].append((case, delta))
with open("avg attn pattern.jsonl", "w") as f:
    import json
    json.dump(jsond, f)
```
